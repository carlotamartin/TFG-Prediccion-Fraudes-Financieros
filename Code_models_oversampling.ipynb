# Análisis de Balanceo de Clases

Debido a que tenemos un desbalance de clases en la variable objetivo, siendo 98.44904% del total del dataset no fraude y el 1.55096% es fraude. Para poder predecir bien el dataset vamos a utilizar la técnica SMOTE (Synthetic Minority Over-sampling Technique).

Se trata de una técnica de sobre-muestreo que se utiliza comúnmente en probemas de clasificación desequilibrados para poder equilibrar la distribución de clases. En nuestor caso, la variable objetivo, que representa si una transacción bancaria es fraude o no, tiene un desbalance con las variables que son fraude.

SMOTE aborda este problema generando nuevos ejemplos sintéticos de la clase minoritaria. Lo hace tomando ejemplos de la clase minoritaria y generando ejemplos sintéticos "interpolando" entre ellos. Básicamente, SMOTE selecciona ejemplos de la clase minoritaria y calcula los k vecinos más cercanos (k-nearest neighbors) para cada ejemplo. Luego, crea nuevos ejemplos sintéticos entre el ejemplo seleccionado y sus vecinos cercanos.

```python
# Definir el pipeline con SMOTE y RandomUnderSampler
pipeline = Pipeline([
    ('over', SMOTE(sampling_strategy=0.1)),  # Ajusta el valor de sampling_strategy según lo necesites
    ('under', RandomUnderSampler(sampling_strategy=0.5))  # Ajusta el valor de sampling_strategy según lo necesites
])
```

```python
# Aplicar el pipeline
X_train_balanced, y_train_balanced = pipeline.fit_resample(X_train, y_train)
```

```python
antes_balanceo = y_train.value_counts()
despues_balanceo = y_train_balanced.value_counts()
print(f'Antes del balanceo', antes_balanceo)
print(f'Después del balanceo', despues_balanceo)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Suponiendo que 'antes_balanceo' y 'despues_balanceo' son series con índices de clases
clases = antes_balanceo.index
antes = antes_balanceo.values
despues = despues_balanceo.values

ind = np.arange(len(clases))  # las ubicaciones de las etiquetas de las clases
width = 0.35  # el ancho de las barras

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, antes, width, label='Antes del balanceo')
rects2 = ax.bar(ind + width/2, despues, width, label='Después del balanceo')

# Añadir texto para etiquetas, título y etiquetas de ejes personalizadas
ax.set_xlabel('Clases')
ax.set_ylabel('Número de instancias')
ax.set_title('Comparación de la Distribución de Clases Antes y Después del Balanceo')
ax.set_xticks(ind)
ax.set_xticklabels(clases)
ax.legend()

# Función para añadir etiquetas sobre cada barra
def autolabel(rects):
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 puntos verticales de desplazamiento
                    textcoords="offset points",
                    ha='center', va='bottom')

autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

### Función curva ROC

```python
def plot_roc_curve(model, X_test, y_test):
    """
    Esta función grafica la curva ROC para un modelo dado y conjunto de prueba.

    Args:
    model: modelo entrenado de scikit-learn.
    X_test: numpy array o dataframe de pandas con los datos de prueba.
    y_test: array o serie de pandas con las etiquetas verdaderas.
    """
    # Obtener las probabilidades de las clases
    y_scores = model.predict_proba(X_test)[:, 1]

    # Calcular FPR, TPR, y umbrales
    fpr, tpr, thresholds = roc_curve(y_test, y_scores)

    # Calcular el AUC
    roc_auc = auc(fpr, tpr)

    # Crear la figura para el plot
    plt.figure()
    plt.plot(fpr, tpr, color='darkorange', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Tasa de Falsos Positivos')
    plt.ylabel('Tasa de Verdaderos Positivos')
    plt.legend(loc="lower right", title='Leyenda')
    plt.show()

    return roc_auc  # Opcional, retorna el AUC en caso de que se quiera utilizar
```

## MLP for Classification

```python
MLP_Clas = Sequential()
# Añadimos capas ocultas
MLP_Clas.add(Dense(10, input_shape=(103, ), activation='sigmoid'))
MLP_Clas.add(BatchNormalization())

MLP_Clas.add(Dense(16, activation='sigmoid'))
MLP_Clas.add(BatchNormalization())
# MLP_Clas.add(Dropout(0.2))

MLP_Clas.add(Dense(10, activation='sigmoid'))
# Output layer
MLP_Clas.add(Dense(1, activation='sigmoid'))

# Learning rate and optimizer
learning_rate = 0.1
optimizer = Adam(learning_rate=learning_rate)

# compile the keras model
MLP_Clas.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy', Recall()]) # metrics=['accuracy']

history_clas = MLP_Clas.fit(X_train_scaled, y_train, validation_data=(X_test_scaled, y_test), epochs=90, batch_size=30)
# evaluate the keras model
_, accuracy_tr, recall_tr = MLP_Clas.evaluate(X_train_scaled, y_train)
_, accuracy_ts, recall_ts = MLP_Clas.evaluate(X_test_scaled, y_test)
print('Accuracy train: %.2f; Recall train: %.2f' % (accuracy_tr*100, recall_tr*100))
print('Accuracy test: %.2f; Recall test: %.2f' % (accuracy_ts*100, recall_ts*100))
```
